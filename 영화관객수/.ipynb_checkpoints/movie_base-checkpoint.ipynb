## 데이터 및 라이브러리 로딩

```python
import pandas as pd  # 데이터프레임 처리 라이브러리
from sklearn.ensemble import RandomForestRegressor  # 랜덤 포레스트 회귀 모델
from sklearn.metrics import mean_squared_error  # 평균 제곱 오차(MSE) 평가 함수
import xgboost as xgb  # XGBoost 모델 라이브러리
import numpy as np  # 수치 연산 라이브러리

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
submission = pd.read_csv('submission.csv')
```

## 데이터 전처리

```python
# 결측치가 많은 데이터 제거
train = train.drop(['dir_prev_bfnum'],axis = 1)
test =  test.drop(['dir_prev_bfnum'],axis = 1)
```

```python
# 감독명 : 너무 다양한 항목을 보유
train = train.drop(['director'],axis = 1)
test = test.drop(['director'],axis = 1)
```

```python
# 제목 : 제목이 관객수와 연관 여부 X
train = train.drop(['title'],axis= 1)
test = test.drop(['title'],axis= 1)
```

```python
train.distributor.value_counts()
```

```python
# 상위 5개의 배급사를 제외하고 '기타'로처리
distributor_list = train.distributor.value_counts()[:5]
def func(distributor):
    if distributor in distributor_list:
        return distributor
    else:
        return '기타'

train['distributor'] = train['distributor'].apply(func)
test['distributor'] = test['distributor'].apply(func)
```

```python
def movie_year(release_time):
    return int(release_time[:4])

def movie_month(release_time):
    return int(release_time[5:7])

train['년'] = train['release_time'].apply(movie_year)
test['년'] = test['release_time'].apply(movie_year)

train['월'] = train['release_time'].apply(movie_month)
test['월'] = test['release_time'].apply(movie_month)

# release_time 열 제거
train = train.drop(['release_time'], axis=1)
test = test.drop(['release_time'], axis=1)
```

```python
# 원핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

## 모델정의 및 학습

```python
train_x = train.drop(['box_off_num'],axis= 1)
train_y = train['box_off_num']
```

```python
model=RandomForestRegressor(n_estimators=100)
model.fit(train_x,train_y)
```

```python
model = xgb.XGBRegressor(n_estimators=10)
model.fit(train_x, train_y)
```

```python
# RMSE 함수 정의
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# 훈련 데이터 성능 평가
train_pred = model.predict(train_x)  # 훈련 데이터 예측값
train_rmse = rmse(train_y, train_pred)
print("Train RMSE:", train_rmse)
```

# 학습 된 모델로 예측 데이터 생성

```python
pred = model.predict(test)

# predict() 메서드는 입력된 특성 값들에 대해 예측된 결과(타겟 값의 추정값)를 반환합니다.
```

## 제출파일 생성

```python
submission['box_off_num'] = pred
```
